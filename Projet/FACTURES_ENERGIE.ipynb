```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("Invoice_20201220.csv", encoding='latin-1', sep="|")
data
```

```python
#affichage colonne site_name dans le format 3chiffres et deux lettres

data['SITE_NAME'] = data['SITE_NAME'].str[11:17]

print(data)
```

```python
#suppression colonne INVOICE_NAME
data = data.drop(columns="INVOICE_NAME")
print(data)
```

```python
#ajout colonne calcul jours inclus sur invoice
#to_datetime : convertit série en objet date
data['END'] = pd.to_datetime(data["END"])
data['BEGIN'] = pd.to_datetime(data['BEGIN'])



#dt.days = nombre de jours pour chaque élément de la série
#pd.series permet d'ajouter la serie en colonne à la dataframe

data['DAYS_INVOICE'] = pd.Series( ((data['END'] - data['BEGIN']).dt.days)+1 )
print(data)
```

```python
#création d'une nouvelle colonne KWH  

#on supprime les espaces
data['KWH_NEW'] = data['KWH_PERIOD'].str[0:1]+data['KWH_PERIOD'].str[2:]

#on transforme l'objet en float
data['KWH_NEW']= pd.to_numeric(data['KWH_NEW'], errors='coerce')
print(data)
```

```python
#suppression colonne KWH_PERIOD
data = data.drop(columns="KWH_PERIOD")
print(data)
```

```python
#calcul de la consommation journalière d'énergie grâce à la nouvelle colonne KWH
DAYLY_AVERAGE_CONSUMPTION = (data['KWH_NEW']/data['DAYS_INVOICE']).round(2)
data['DAYLY_AVERAGE_CONSUMPTION']=pd.Series(DAYLY_AVERAGE_CONSUMPTION)

print(data)
```

```python
#creation d'un nouveau tableau présentant le nombre d'invoices par antenne
NUMBER_OF_INVOICES = (data.groupby(by ='SITE_NAME').size().reset_index(name='INVOICES'))
data_2 = pd.DataFrame(NUMBER_OF_INVOICES)
print(data_2)
```

```python
#fusion des tableaux data et data_2, but :  associer le nombre d'invoices à chaque antenne sur le tableau de base
data_new = pd.merge(left=data, right=data_2, how="left", left_on="SITE_NAME", right_on="SITE_NAME")
data_new
```
